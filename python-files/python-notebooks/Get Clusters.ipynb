```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from  sklearn.neighbors import KNeighborsClassifier
import numpy as np
from shapely.geometry import Point
from geopandas import GeoDataFrame
import geopandas as gpd
from fiona.crs import from_epsg
import matplotlib.pyplot as plt
import os
from os.path import join
from glob import iglob
```

```python
base_map_gdf = gpd.read_file('Demographics_of_India.dbf')
```

```python
directory = 'Hotspots'
files = sorted(list(iglob(join(directory,'*.csv'),recursive=True)))
```

```python
all_cluster_center = []

for filename in files:
    data = pd.read_csv(filename)
    clustering_class = feb_cls = AgglomerativeClustering(n_clusters=None, compute_full_tree=True, distance_threshold=5)
    clustering_class.fit(data[['latitude', 'longitude']])
    data = data.assign(cls=clustering_class.labels_)
    cluster_center = []
    for label in np.unique(clustering_class.labels_):
        temp =data[data['cls']== label]
        cluster_center.append((temp.latitude.mean(),temp.longitude.mean()))
    all_cluster_center.append(cluster_center)
```

```python
info = {}
```

```python
classifier = KNeighborsClassifier(weights='distance')
classifier.fit(all_cluster_center[0], range(len(all_cluster_center[0])))
```

```python
number_of_centers = []
for x in all_cluster_center:
    number_of_centers.append(len(x))
m = np.array(number_of_centers)
temp = m.argsort()[-1]
max_clusters = temp
print(max_clusters)
```

```python
for i,center in enumerate(all_cluster_center[max_clusters]):
    info[i] = [center]
```

```python
for centers in all_cluster_center[1:]:
    cls = classifier.predict(centers)
    for i, c in enumerate(cls):
        info[c] += [centers[i]]
```

```python
arr = []
for labes, centers in info.items():
    for center in centers:
        lat, long = center
        arr.append((lat, long, labes))
```

```python
final = pd.DataFrame(arr, columns=['latitude', 'longitude', 'cluster'])
```

```python
geometry = [Point(xy) for xy in zip(final.longitude, final.latitude)]
crs = {'init': 'epsg:31256'}
final_gdf = GeoDataFrame(final, crs=from_epsg(31256), geometry=geometry)
```

```python
base_map = base_map_gdf.plot(figsize=(20, 20), facecolor='white', edgecolor='grey')
final_gdf[final_gdf['cluster']==2].plot(ax=base_map, column='cluster')
```
