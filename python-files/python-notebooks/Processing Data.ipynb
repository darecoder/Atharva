```python
for i in input_files_OFFL:
#     print("working on", i)
    try:
        harp_L2_L3 = harp.import_product(i, operations =" \
            O3_column_number_density_validity>50; \
            derive(O3_column_number_density); \
            derive(datetime_stop {time}); \
            latitude > 8.4 [degree_north]; latitude < 37.6 [degree_north]; longitude > 68.7 [degree_east]; longitude < 97.2[degree_east]; \
            bin_spatial(292, 8.4, 0.1, 285, 68.7, 0.1); \
            derive(latitude {latitude}); derive(longitude {longitude}); \
            keep(latitude,longitude,O3_column_number_density);")

        export_file = "{export_path}/{name}".format(export_path=export_path, name=i.split('/')[-1].replace('L2','L3'))
        harp.export_product(harp_L2_L3, export_file, file_format='netcdf')
    except:
        print(f'ERROR: {i} was not processed')
        continue
```

```python
attributes = {}
for i in input_files_OFFL:
    try:
        attributes.update({
            i.split('/')[-1]:{
                'time_coverage_start':xr.open_dataset(i).attrs['time_coverage_start'],
                'time_coverage_end':xr.open_dataset(i).attrs['time_coverage_end'],
            } 
        })
    except Exception as e:
        print(e)
        continue
```

```python
filename_L3=sorted(list(iglob(join(export_path,'**','*L3__O3*'),recursive=True)))
def preprocess(ds):
    ds['time']=pd.to_datetime(np.array([attributes[ds.attrs['source_product']]['time_coverage_start']])).values
    return ds
```

```python
filename_L3
```

```python
raw_combined_data=xr.open_mfdataset(filename_L3,combine='nested',concat_dim='time',preprocess=preprocess,chunks={'time':100})
```

```python
raw_combined_data
```

```python
raw_combined_data.to_dataframe()
```

```python
mean_per_day_data = raw_combined_data.resample(time='1D').mean(dim='time',skipna=None)
max_per_day_data = raw_combined_data.resample(time='1D').max(dim='time',skipna=None)
min_per_day_data = raw_combined_data.resample(time='1D').min(dim='time',skipna=None)
```

```python
# Save weekly processed data
processed_data_folder = f'Weekly processed {GAS}'
if not os.path.exists(processed_data_folder):
    os.mkdir(processed_data_folder)
```

```python
mean_week_data = mean_per_day_data.mean(dim='time')
max_week_data = max_per_day_data.max(dim='time')
min_week_data = min_per_day_data.min(dim='time')
```

```python
mean_week_data = mean_week_data.rename_vars(O3_column_number_density='mean_O3_column_number_density')
max_week_data = max_week_data.rename_vars(O3_column_number_density='max_O3_column_number_density')
min_week_data = min_week_data.rename_vars(O3_column_number_density='min_O3_column_number_density')
```

```python
final_data = (mean_week_data.merge(min_week_data)).merge(max_week_data)
```

```python
final_data.to_dataframe().to_csv(os.path.join(processed_data_folder, f'{GAS}_{product_path.split("___")[1]}.csv'))
```
